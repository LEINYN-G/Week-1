```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Activation, Dropout, Flatten, Dense, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
from tensorflow.keras.utils import plot_model
from glob import glob
```

```python
#visualization
import glob
import cv2
import pandas as pd
from tqdm import tqdm  # Import tqdm for progress bar

# Initialize data lists
x_data = []
y_data = []

# Process each category
for category in glob.glob(train_path + '/*'):  # Use glob.glob to ensure proper usage
    for file in tqdm(glob.glob(category + '/*'), desc=f"Processing {category}"):  # Display progress with tqdm
        img_array = cv2.imread(file)
        img_array = cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB)
        x_data.append(img_array)
        y_data.append(category.split('/')[-1])

# Create a DataFrame
data = pd.DataFrame({'image': x_data, 'label': y_data})

# Display a summary of the DataFrame
print(data.head())
```

```python
import matplotlib.pyplot as plt  # Import matplotlib.pyplot

# Define colors for the pie chart
colors = ['#a0d157', '#c48bb8']

# Create the pie chart
plt.pie(
    data['label'].value_counts(),  # Get the value counts for labels
    labels=['Organic', 'Recyclable'],  # Labels for the chart
    colors=colors,  # Assign custom colors
    autopct='%1.1f%%',  # Display percentages
    startangle=90,  # Start pie chart at 90 degrees
    explode=(0.1, 0)  # Slightly separate the first slice
)

# Add a title
plt.title('Distribution of Organic and Recyclable Items')

# Show the pie chart
plt.show()
```

## CNN -- Convolutional Neural Network

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=(224, 224, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D())

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D())

model.add(Conv2D(128, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D())

model.add(Flatten())

model.add(Dense(256))
model.add(Activation('relu'))
model.add(Dropout(0.5))

model.add(Activation('relu'))
model.add(Dropout(0.5))

model.add(Dense(2))
model.add(Activation('sigmoid'))


model.compile(loss = "binary_crossentropy",
              optimizer = "adam",
              metrics = ["accuracy"])
batch_size = 256
```

```python
model.summary()
```

```python
# Define data augmentation settings
train_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
# load taining data
train_generator = train_datagen.flow_from_directory(
    train_path,
    target_size = (224, 224),
    batch_size = batch_size,
    color_mode = "rgb",
    class_mode = "categorical")

#load testing data
test_generator = test_datagen.flow_from_directory(
    test_path,
    target_size = (224, 224),
    batch_size = batch_size,
    color_mode = "rgb",
    class_mode = "categorical")
```

```python
hist = model.fit(
    train_generator,
    epochs = 10,
    validation_data=test_generator
)
```
